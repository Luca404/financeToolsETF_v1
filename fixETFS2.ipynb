```python
import pandas as pd
```

```python
dataMilano = pd.read_csv( "ETF_Milano_1.csv" )
dataMilano
```

```python
dataMilano[["Area","Sector","Maturities"]] = dataMilano["Area"].str.split(",",2,expand=True)
```

```python
dataMilano = dataMilano.drop('info', axis=1)
```

```python
dataMilano
```

```python
dataMilano.to_csv("ETF_Milano_1.csv")
```

```python
dataXetra = pd.read_csv( "ETF_Xetra_1.csv" )
dataXetra
```

```python
dataXetra["Symbol"] = dataXetra["Symbol"] + ":XETRA"
```

```python
dataXetra
```

```python
dataXetra[["Area","Sector","Maturities"]] = dataXetra["Area"].str.split(",",2,expand=True)
dataXetra
```

```python
dataXetra = dataXetra.drop('info', axis=1)
```

```python
etfList = pd.concat([dataMilano, dataXetra], ignore_index=True)
```

```python
etfList
```
